```python
# Function
def greet(name):
    print(f"Aslam o Alaikum, {name}!\n")
```

```python
# Method
class Person:
    def __init__(self, name): # Parameterized Constructor
        self.name = name

    def greet(self):
        print(f"Aslam o Alaikum!\n Ksa han ap, {self.name}.")

# Calling the function
greet("Muhammad Arslan")

# Calling the method
person = Person("Muhammad Arslan")
person.greet()
```

```python
# Non Keyword Arguments & Keyword Arguments
def test_args(name, *args, **kwargs):
    print(f"*args: {args}")
    print(f"**kwargs: {kwargs}")
        
test_args('Muhammad Arslan', "Saad", 1, "go", 2.1, "school", title = "Hashim Malik", stage = 4)
```

```python
# Recursion / Recursive function
# Example 1
def recursive_fibonacci(n):
    if n <= 1:
        return n
    else:
        return(recursive_fibonacci(n-1) + recursive_fibonacci(n-2))
        
b = [i for i in range(10)]

fib_number = [recursive_fibonacci(i) for i in b]
print(fib_number)

# Example 2
def factorial(n):
    if n == 0 or n == 1:
        return 1

    else:
        return n * factorial(n-1)

result = factorial(5)
print(result)
```
